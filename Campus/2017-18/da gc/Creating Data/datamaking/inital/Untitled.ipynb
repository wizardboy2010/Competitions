```python
def numb(file, word):
    with open(file) as f:
        contents = f.read()
        count = contents.count(word)
    return count
```

```python
def reqline(file, word):
    a = []
    with open(file) as f:
        for lines in f:
            if word in lines:
                a.append(lines[:-1])
    return a
```

```python
files = ['Aditya Birla Sun Life Mutual Fund.txt',
         'HDFC Mutual Fund.txt',
         'ICICI Prudential Mutual Fund.txt',
         'Sundaram Mutual Fund.txt',
         'Reliance Mutual Fund.txt',
         'SBI Mutual Fund.txt',
         'Tata Mutual Fund.txt',
         'UTI Mutual Fund.txt']
```

```python
for f in files:
    print(numb(f, 'Open Ended Schemes'))
```

```python
a = np.array([])
for f in files:
    #print(f)
    a = np.append(a, reqline(f, 'Open Ended Schemes'))
funds = np.unique(a)
del a
print(funds)
```

```python
schemes = []
for f in files:
    d = pd.read_csv(f, delimiter=';', low_memory=False)
    schemes.append(np.unique(d['Scheme Name'][~d['Scheme Name'].isnull()]))
```

```python
pkl.dump(schemes, open("schemes.pkl", "wb"))
```

```python
c = 0
for i in schemes:
    c += len(i)
c
```
