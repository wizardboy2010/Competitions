```python
def count(file, word):
    with open(file) as f:
        contents = f.read()
        count = contents.count(word)
    return count
```

```python
def line(file, word):
    a = []
    with open(file) as f:
        for lines in f:
            if word in lines:
                a.append(lines[:-1])
    return a
```

```python
for f in files:
    print(count(f, 'Open Ended Schemes'))
```

#### Data Extraction for diff funds

```python
col = ['Scheme Code', 'Scheme Name', 'Net Asset Value', 'Repurchase Price', 'Sale Price', 'Date']
```

```python
def funddata(file, fundname):
    a = []
    flag = 0
    start = 0
    with open(file) as f:
        for lines in f:
            if fundname in lines[:-1]:
                flag = 1
            if lines[:-1] == file[:-4]:
                start = 1
            if lines == '\n':
                start = 0
            if flag ==1 and start ==1:
                a.append(lines)
    return a
```
